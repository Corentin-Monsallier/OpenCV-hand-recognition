## 2. Project Name

````python
def main():
    ####################          CAMERA SETTINGS          ####################
    pTime, cTime = 0, 0
    cap = cv2.VideoCapture(0)
    cap.set(3, 1280)
    cap.set(4, 720)

    ####################          FACE DETECTOR CONFIG          ####################
    detector = faceDetector(maxFaces=2) # Adjust maxFaces as needed

    while True:
        ####################          ESC: end the program          ####################
        key = cv2.waitKey(10)
        if key == 27:
            break

        ####################          FACE DISPLAY          ####################
        success, img = cap.read()
        if not success:
            break
        img = detector.findFaceMesh(img)
        lmList = detector.findPosition(img, multiple=True)

        ####################          FPS DISPLAY          ####################
        cTime = time.time()
        fps = 1 / (cTime - pTime)
        pTime = cTime
        cv2.putText(img, f'FPS: {int(fps)}', (10, 70), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3)

        ####################          DISPLAY IMAGE          ####################
        cv2.imshow("Image", img)
        cv2.waitKey(1)
````

````python
if __name__ == "__main__":
    main()
````
